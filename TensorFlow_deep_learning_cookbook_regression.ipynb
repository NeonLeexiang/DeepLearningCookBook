```python
# import tensorFlow
import tensorflow as tf

```

```python
# the dimension of x and w1
m = 1    # the amount of sample
n = 1      # the amount of feature

# placeholder for training data
x = tf.placeholder(tf.float32, name='x', shape=[m, n])
y = tf.placeholder(tf.float32, name='y')

w0 = tf.Variable(0.0)
w1 = tf.Variable(tf.random_normal([n, 1]))

# the linear regression model
y_hat = tf.matmul(x, w1) + w0

# Multiple linear regression loss
loss = tf.reduce_mean(tf.square(y - y_hat), name='loss')

```

```python
# some error with the regression
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train_step = optimizer.minimize(loss)

x_data = []
y_data = []

with tf.Session() as sess:
    rst = sess.run(train_step, feed_dict={x:x_data, y:y_data})
    print(rst)
    

```
